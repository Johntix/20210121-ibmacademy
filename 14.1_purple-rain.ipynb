## 2. Ajout d'une méthode `deplace()` qui permet à la goutte de "tomber"

`canvas.move(element,dx,dy)`
`root.after(ms,callback)`

```python
#------------------------------------------------------------------#
# Purple Rain (sur une idée de Daniel Schiffman, The Coding Train) # 
# 2 ! goutte tombe                                             #
#------------------------------------------------------------------#
from tkinter import *

Largeur = 800
Hauteur = 600

class Drop(object):
    def __init__(self, x, y, v):
        self.x = x
        self.y = y
        self.v = v
        self.element = ''
 
    def affiche(self):
        self.element = canvas.create_rectangle(self.x-1,self.y-5,self.x+1,self.y+5, fill ="pink", width=0)
   
    def deplace(self):
        canvas.move(self.element,0,self.v)
        root.after(5,self.deplace)

# Fenêtre tkinter
root = Tk()    
canvas = Canvas(root, width=Largeur, height=Hauteur, background="purple")
canvas.pack(side=LEFT, padx=5, pady=5)

root.title("Purple Rain 2")

goutte = Drop(250,600,-1)
goutte.affiche()
goutte.deplace()

goutte1 = Drop(650,0,2)
goutte1.affiche()
goutte1.deplace()


root.mainloop()
```

## 3. Dans la méthode `deplace()` on gère la hauteur de la goutte et on la fait repartir du haut lorsqu'elle a atteint le "sol" (la hauteur du canvas)


```python
#------------------------------------------------------------------#
# Purple Rain (sur une idée de Daniel Schiffman, The Coding Train) # 
# 2 ! goutte tombe                                             #
#------------------------------------------------------------------#
from tkinter import *
from random import randint

Largeur = 800
Hauteur = 600

class Drop(object):
    def __init__(self, x, y, v):
        self.x = x
        self.y = y
        self.v = v
        self.element = ''
 
    def affiche(self):
        self.element = canvas.create_rectangle(self.x-1,self.y-5,self.x+1,self.y+5, fill ="pink", width=0)
   
    def deplace(self):
        canvas.move(self.element,0,self.v)
        root.after(5,self.deplace)
        self.y = self.y + self.v
        if self.y > Hauteur:
            self.y = 0
            canvas.move(self.element,0, - Hauteur)

# Fenêtre tkinter
root = Tk()    
canvas = Canvas(root, width=Largeur, height=Hauteur, background="purple")
canvas.pack(side=LEFT, padx=5, pady=5)

root.title("Purple Rain 2")

goutte = Drop(randint(0,Largeur), randint(-50, -25), randint(1,4))
goutte.affiche()
goutte.deplace()


root.mainloop()
```

## 4. On fait démarrer la goutte à une position latérale et verticale random, plus haut que le canvas (valeur négative de y), et avec une vitesse de chute aléatoire.


## 5. Création d'une classe `Rain()` pour gérer un ensemble de gouttes  


```python
#------------------------------------------------------------------#
# Purple Rain (sur une idée de Daniel Schiffman, The Coding Train) # 
# 2 ! goutte tombe                                             #
#------------------------------------------------------------------#
from tkinter import *
from random import randint

Largeur = 800
Hauteur = 600

class Drop(object):
    def __init__(self, x, y, v):
        self.x = x
        self.y = y
        self.v = v
        self.element = ''
 
    def affiche(self):
        self.element = canvas.create_rectangle(self.x-1,self.y-5,self.x+1,self.y+5, fill ="pink", width=0)
   
    def deplace(self):
        canvas.move(self.element,0,self.v)
        root.after(5,self.deplace)
        self.y = self.y + self.v
        if self.y > Hauteur:
            self.y = randint(-50, -25)
            self.x = randint(0,Largeur)
            self.v = randint(1,4)
            self.element = canvas.create_rectangle(self.x-1,self.y-5,self.x+1,self.y+5, fill ="pink", width=0)

class Rain():
    def __init__(self, force):
        self.force = force
        self.goutte = []
        for i in range(self.force):
            self.goutte.append(Drop(randint(0,Largeur), randint(-50, -25), randint(1,4)))
        
    def tombe(self):
        for i in range(len(self.goutte)):
            self.goutte[i].affiche()
            self.goutte[i].deplace()

# Fenêtre tkinter
root = Tk()    
canvas = Canvas(root, width=Largeur, height=Hauteur, background="purple")
canvas.pack(side=LEFT, padx=5, pady=5)

root.title("Purple Rain 5")

pluie = Rain(60)
pluie.tombe()

root.mainloop()
```

## 6. Pour finir : les gouttes qui tombent le plus vite sont plus grosses que les autres (pour créer une impression de profondeur)


```python
#------------------------------------------------------------------#
# Purple Rain (sur une idée de Daniel Schiffman, The Coding Train) # 
# 2 ! goutte tombe                                             #
#------------------------------------------------------------------#
from tkinter import *
from random import randint
from math import floor 

Largeur = 800
Hauteur = 600

class Drop(object):
    def __init__(self, x, y, v):
        self.x = x
        self.y = y
        self.v = v
        self.element = ''
 
    def affiche(self):
        self.element = canvas.create_rectangle(self.x-1- self.v,self.y-5 -3 * self.v,self.x+1,self.y+5, fill ="pink", width=0)
    def deplace(self):
        canvas.move(self.element,0,self.v)
        root.after(5,self.deplace)
        self.y = self.y + self.v
        if self.y > Hauteur:
            self.y = randint(-50, -25)
            self.x = randint(0,Largeur)
            self.v = randint(1,4)
            self.element = canvas.create_rectangle(self.x-1-self.v,self.y-5 - 3 * self.v,self.x+1,self.y+5, fill ="pink", width=0)

class Rain():
    def __init__(self, force):
        self.force = force
        self.goutte = []
        for i in range(self.force):
            self.goutte.append(Drop(randint(0,Largeur), randint(-50, -25), randint(1,4)))
        
    def tombe(self):
        for i in range(len(self.goutte)):
            self.goutte[i].affiche()
            self.goutte[i].deplace()

# Fenêtre tkinter
root = Tk()    
canvas = Canvas(root, width=Largeur, height=Hauteur, background="purple")
canvas.pack(side=LEFT, padx=5, pady=5)

root.title("Purple Rain 5")

pluie = Rain(60)
pluie.tombe()

root.mainloop()
```
