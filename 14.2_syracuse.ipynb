Ecrire une fonction qui pour un entier donné applique l'agorithme suivant: 
  
La suite de Syracuse fait l'objet d'une conjecture en mathématiques, on suppose, mais l'on ne sait pas si elle converge toujours vers $1$, elle est définie comme suit : 

On démarre avec un entier $n$ : 

- si $n$ est pair : $n \leftarrow n/2$ 
- si $n$ est impair : $n \leftarrow 3n+1$
- quand $n$ vaut $1$ : fin

On demande : 

- d'afficher la liste des valeurs de la suite de Syracuse pour un entier $n$, 
- ajouter le calculer du "temps de vol" (nombre d'itérations pour arriver à 1) 
- ajouter le calcul de son "altitude" (plus haute valeur atteinte).


- trouver le temps de vol maximal pour les valeurs dans l'intervalle $[m,n]$
- grapher le temps de vol  pour les valeurs de $n$ dans l'intervalle $[m,n]$ 
- trouver l'altitude maximale pour les suites de $m$ à $n$
- grapher l'altitude pour les suites de $m$ à $n$



```python
def syracuse(n):
    temps_de_vol = 0
    altitude = 0
    while n != 1:
        if n > altitude:
            altitude = n
        temps_de_vol += 1 
        if n % 2 == 0: 
            n = n / 2 
        else: 
            n = 3 * n + 1 
    return altitude,temps_de_vol

print(syracuse(2425))
```

```python
def intervalle(m,n):
    alts = []
    tvol = []
    for i in range(m,n):
        a,t = syracuse(i)
        alts.append(a)
        tvol.append(t)
    return alts,tvol
```

```python
m, n = 1, 2500

altitudes, temps = intervalle(m,n)

# il reste juste à plotter ces 2 arrays
import matplotlib.pyplot as plt

x = [i for i in range(m,n)]

plt.scatter(x,temps)
plt.show()

plt.scatter(x,altitudes)
plt.show()
```
