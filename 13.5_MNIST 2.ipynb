```python
import pandas as pd
```

```python
df_train = pd.read_csv('mnist_train.csv')
df_test = pd.read_csv('mnist_test.csv')
df_test.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_test.iloc[:,1:] #pour selectionner toutes les colones sauf la premiere 'label' qui sera la cible
y = df_test['label']
```

```python
X_train,X_test, y_train,y_test= train_test_split(X,y, test_size=0.3,random_state=42, stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier as dt
from sklearn.linear_model import LogisticRegression as lr
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.ensemble import RandomForestClassifier as rf
model_1 = dt(random_state=42)
model_2 = knn(n_neighbors = 2, n_jobs = -1)
model_3 = lr()
model_4 = rf(n_estimators= 50, verbose = 2, n_jobs = -1)
model_4.get_params()
```

```python
for model in(model_2,model_4) : 
    model.fit(X_train, y_train)
    print(model.__class__.__name__, model.score(X_test, y_test))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
'max_depth': [2,15],  
'max_features': range(27,42),   
'n_estimators': [10,42],
}
```

```python
grid = GridSearchCV(model_4, param_grid, cv=10, scoring='accuracy')
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
best_rf = grid.best_estimator_
```

```python
Y_pred = best_rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,Y_pred))
```
