<h1 align=center>Visualiser les données Chipotle</h1>

### 1. Les imports

```python
import pandas as pd
import matplotlib.pyplot as plt 

%matplotlib inline
```

### 2. Reprendre le jeu de données chipotle.tsv
Créer un dataFrame *chipo*

```python
chipo = pd.read_csv('chipotle.tsv', sep='\t')
```

### 3. Créer un histogramme représentant les 5 produits les plus vendus

```python
#Top 5
df=chipo.item_name.value_counts()[0:5]

# Histogramme
df.plot(kind='bar')

# Légendes
plt.xlabel('Produits')
plt.ylabel('Prix')
plt.title('Top 5 des produits commandés')

#Affichage
plt.show()
```

### 4. Créer un nuage de points représentant le nombre d'articles commandés en fonction du montant de la commande

```python
# Prix sans le $
#chipo.item_price = [float(value[1:-1]) for value in chipo.item_price] 

# groupes par commandes
commandes = chipo.groupby('order_id').sum()

# Nuage de points
plt.scatter(x = commandes.item_price, y = commandes.quantity, s = 50, c = 'green')

# Set the title and labels
plt.xlabel('Prix de la commande')
plt.ylabel('Produits command&s')
plt.title('Nombre d\'articles / Prix de la commande' )
plt.ylim(0,10)
```
