<h1 align=center>Utilisation de Seaborn</h1>

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%pip install seaborn==0.11.0
```

### 2. Créer un dataFrame *tips* à partir du fichier *tips.csv*

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

### 3. Visualiser les premières lignes

### 4. Supprimer la colonne Unnamed 0

```python
tips.drop(['Unnamed: 0'], axis=1, inplace=True)
tips.head()
```

### 5. Représenter la colonne total_bill par un histogramme

```python
test1 = sns.distplot(tips.total_bill)
test1
```

### 6. Créer un nuage de points représentant la relation entre les colonnes total_bill et tip

```python
#tips['color']=tips['total_bill'].apply(lambda x : 'Blue' if x=='male' else 'Red')

#nuage = tips.plot.scatter('tip', 'total_bill', c='color')

# titre
#nuage.set(title = 'total_bill / tip')

test2 = sns.jointplot(x='total_bill',y='tip', data= tips)
```

### 7.  Créer une image avec la relation entre total_bill, tip et size.

```python
#tips['color']=tips['total_bill'].apply(lambda x : 'Blue' if x=='male' else 'Red')

#nuage = tips.plot.scatter('size', 'total_bill', c='color')

# titre
#nuage.set(title = 'total_bill / size')

test_3 = sns.pairplot(tips)
test_3.savefig('output.png')
```

### 8. Présenter la relation entre days et total_bill value

```python
test4 = sns.stripplot(x='day',y='total_bill', data = tips, jitter = True)
```

### 9. Creér les nuages de points avec *day* en ordonnés et *tip* en abscisses, différentier les points par sexe

```python
test5 = sns.stripplot(x = 'day', y = 'total_bill',data =  tips, jitter = True)
test5
```

### 10.  Créer les boites à moustache représentant total_bill par jour en différenciant par moment (Dinner ou Lunch)

```python
test6 = sns.boxplot(x=tips.total_bill,y=tips.day,hue=tips.time)
```

### 11. Créer deux histogrammes représentant , cote à cote, tip value selon Dinner ou Lunch.

```python
test7= sns.FacetGrid(tips, col="time")
test7.map(sns.distplot, "tip")
```

### 12. Créer deux nuages de points, un pour les hommes, l'autre pour les femmes représentants la relation entre total_bill value et tip, en différenciant les points par fumeur ou non fumeur.

```python
g = sns.FacetGrid(tips, col = 'sex', hue = 'smoker')
g.map(plt.scatter, 'total_bill','tip',alpha = .7) #le alpha représente la transparence des points 0.7 équivaut a 70%
g.add_legend()
```
